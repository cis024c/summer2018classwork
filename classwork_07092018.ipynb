##### Map 

```python
def square(x):
    return x*x
```

```python
square(2)
```

```python
numbers = [1,2,3,4,5,6]
```

```python
tuple(map(square,numbers))
```

#### Filter

```python
def odd(x):
    if x%2 == 1:
        return True
    else:
        return False
```

```python
odd(7)
```

```python
odd(6)
```

```python
list(filter(odd,numbers))
```

#### Reduce

```python
import functools

numbers = [1]
```

```python
def sum2(x,y):
    print(x,y)
    y = x + y
    return y
```

```python
functools.reduce(sum2,numbers)
```

```python
total = 0
for i in numbers:
    total = total + i
print(total)
```
