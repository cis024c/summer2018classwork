```python
movieFile = open("movies.csv",'rb')
```

```python
movieFile.seek(0)
movieFile.readline()

movies = []
for line in movieFile:
    movieData = str(line).split(',')
    movies.append(movieData)
```

```python
len(movies)
```

```python
movies[0]
```

```python
movies[:2]
```

```python
movies[1][2].strip("\\r\\n'")
```

```python
genreCount = {}
for movie in movies:
    genres = movie[-1].strip("\\r\\n''").strip('\\r\\n"').split("|")
    for genre in genres:
        if genreCount.get(genre) == None:
            genreCount[genre] = 1
        else:
            genreCount[genre] += 1
genreCount
```

```python
moviesByGenre = {}
for movie in movies:
    genres = movie[-1].strip("\\r\\n''").strip('\\r\\n"').split("|")
    for genre in genres:
        if moviesByGenre.get(genre) == None:
            moviesByGenre[genre] = [movie[0]]
        else:
            moviesByGenre[genre].append(movie[1])
moviesByGenre
```
